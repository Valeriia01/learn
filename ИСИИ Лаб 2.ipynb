2. Показать список уникальных наименований документов, требованиям которых соответствуют представленные средства защиты информации

```python
task2 = data['Наименования документов, требованиям которых соответствует средство'].unique()
uniq = []
for i in task2:
    str_ = i.rsplit(': ')
    if len(str_)>1: str_ = str_[1].rsplit(', ')
    for j in range (len(str_)):
        if str_[j] not in uniq:
            uniq.append(str_[j])
uniq
```

3. Отобразить список наименований средств, соответствующих двум любым документам с требованиями по сертификации

```python
from random import choice

def task3(uniq, data):
    rd = choice(uniq)
    docs = [] 
    for i in range (len(data)):
        if rd in data['Наименования документов, требованиям которых соответствует средство'][i]:
            docs.append(data['Наименование\r\nсредства (шифр)'][i])
    return rd, docs
```

```python
rez = task3(uniq, data)
print(rez[0], ': \n', rez[1])
```

```python
rez = task3(uniq, data)
print(rez[0], ': \n', rez[1])
```

4. Показать наименования тех средств, которые имеют просроченные сертификаты (на момент выполнения задания)

```python
import datetime 
def task4(data):
    rez = []
    for i in range (len(data)):
        if data['Срок\r\nдействия\r\nсертификата'][i] != '#Н/Д':
            doc_date = datetime.datetime.strptime(data['Срок\r\nдействия\r\nсертификата'][i], '%d.%m.%Y').date()
            today = datetime.date.today()
            if today > doc_date and data['Наименование\r\nсредства (шифр)'][i] not in rez :
                rez.append(data['Наименование\r\nсредства (шифр)'][i])
    return rez
```

```python
overdue = task4(data)
overdue
```

5. Показать наименования тех средств, которые имеют просроченные сертификаты, но при этом обязаны оказывать поддержку (столбец «Информация об окончании срока технической поддержки, полученная от заявителя»)

```python
import numpy as np
def task5(data, overdue): 
    rez = []
    df = data['Информация об окончании срока технической поддержки, полученная от заявителя'].fillna('1.01.2000')
    for i in range (len(data)):
        if df[i] == 'бессрочно' : df[i] = '1.01.2100'
        if (data['Наименование\r\nсредства (шифр)'][i] in overdue) and (datetime.datetime.strptime(df[i], '%d.%m.%Y').date() > datetime.date.today()): 
            rez.append(data['Наименование\r\nсредства (шифр)'][i])           
    return rez
```

```python
task5(data, overdue)
```

6. Показать наименования всех средств, которые проходили испытания в частных организациях (форма организации ООО, ЗАО, ПАО, АО)

```python
def task6 (data):
    rez = []
    for i in range (len(data)):
        x = data['Испытательная\r\nлаборатория'][i]
        if 'ООО' in x or 'ЗАО' in x or 'ПАО' in x or 'АО' in x:
            if data['Наименование\r\nсредства (шифр)'][i] not in rez:
                rez.append(data['Наименование\r\nсредства (шифр)'][i])
    return rez
```
