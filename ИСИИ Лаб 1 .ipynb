2. Выполнить генерацию линейного массива (вектора) чисел с плавающей точкой произвольного размера; для каждого массива выполнить оценку времени, затраченного на генерацию; сравнить показатели с массивами, генерируемыми средствами стандартной библиотеки языка Python аналогичной размерности 

```python
def arr_1x(): #генерация линейного массива 
    mas1x = [] 
    size_ = random.randint(100, 10**6) #определение размерности
    print('Длина массива - ', size_)
    print('----------БИБЛИОТЕКА RANDOM----------')
    start_time = datetime.now()
    for i in range (size_):
        mas1x.append(random.uniform(0, 10**7))
    print('Затраченное время ', datetime.now()-start_time) #Вычисление длительности
    print('Первые 5 элементов массива: \n', np.array(mas1x[:5]))
    print('----------БИБЛИОТЕКА NUMPY-----------')
    start_time = datetime.now()
    mas1x_np = np.random.uniform(0, random.randint(100, 10**5), size = size_)
    print('Затраченное время ', datetime.now()-start_time) #Вычисление длительности
    print('Первые 5 элементов массива: \n', mas1x_np[:5])
    print('\n')
```

```python
for i in range (5):
  arr_1x()
```

2. Выполнить генерацию двумерного массива (матрицы) чисел с плавающей точкой произвольного размера; для каждого массива выполнить оценку времени, затраченного на генерацию; сравнить показатели с массивами, генерируемыми средствами стандартной библиотеки языка Python аналогичной размерности; выполнить все арифметические операции над элементами

```python
def arr_2x(): #генерация матрицы
    mas2x = [] 
    #определение размерности
    size_1 = random.randint(100, 10**4)
    size_2 = random.randint(10, 100)
    mas2x = []
    print('Длина массива - ', size_1, '*', size_2)
    print('----------БИБЛИОТЕКА RANDOM----------')
    start_time = datetime.now()
    for i in range (size_1):
      mas2x.append([])
      for j in range (size_2):
        mas2x[i].append(random.uniform(0, 10**7))
    print('Затраченное время ', datetime.now()-start_time) #Вычисление длительности
    print('mas2x[0]: \n', np.array(mas2x[0]))
    print('----------БИБЛИОТЕКА NUMPY-----------')
    start_time = datetime.now()
    mas2x_np = np.random.uniform(0, random.randint(100, 10**5), size = (size_1, size_2))
    print('Затраченное время ', datetime.now()-start_time) #Вычисление длительности
    print('mas2x_np[0]: \n', mas2x_np[0])
    print('\n')
    #Арифметические операции над матрицами
    print('----------ПОЭЛЕМЕНТНОЕ СЛОЖЕНИЕ МАТРИЦ-----------')
    print (mas2x_np[0][:5] + mas2x[0][:5])
    print('----------ПОЭЛЕМЕНТНОЕ УМНОЖЕНИЕ МАТРИЦ-----------')
    print (mas2x_np[0][:5] * mas2x[0][:5])
    print('----------ПОЭЛЕМЕНТНОЕ ВЫЧИТАНИЕ МАТРИЦ-----------')
    print (mas2x_np[0][:5] - mas2x[0][:5])
    print('----------ПОЭЛЕМЕНТНОЕ ДЕЛЕНИЕ МАТРИЦ-----------')
    print (mas2x_np[0][:5] / mas2x[0][:5])
    print('----------ПРИБАВЛЕНИЕ ЗНАЧЕНИЯ-----------')
    print (mas2x_np[0][:5] + 5)
    print('----------УМНОЖЕНИЕ НА ЧИСЛО-----------')
    print (mas2x_np[0][:5] * 5)
    print('----------ВЫЧИТАНИЕ ЧИСЛА-----------')
    print (mas2x_np[0][:5] - 5)
    print('----------ДЕЛЕНИЕ НА ЧИСЛО-----------')
    print (mas2x_np[0][:5] / 5)
    print('\n')
```

```python
for i in range (2):
  arr_2x()
```

3. Реализовать игру «Жизнь» c использованием NumPy

```python
# Константы цветов RGB
BLACK = (0 , 0 , 0)
WHITE = (255 , 255 , 255)
# Создаем окно
root = p.display.set_mode((640 , 640))

# "Жители" на поле 10*10 (0 - мертв, 1 - жив )
cells = np.random.choice([0,1], (10, 10))

DO = 1
data = []
# Основной цикл
while DO:
    data.append(np.array(cells)) #храним шаги 
    
    # Отрисовка
    for i in range(len(cells)):
        for j in range(len(cells[0])):
            p.draw.rect(root , (255 * cells[i][j] % 256 , 0 , 0) , [i * 64 , j * 64 , 64 , 64])
    
    # Обновляем экран
    p.display.update()
    time.sleep(0.1) #чтобы глаза не болели от скорости смены картинки
    for i in range (len(cells)):
        for j in range (len(cells[0])):
            #Учет циклического поля
            k = i - 1
            m = i + 1
            if i == 0: k = len(cells)-1
            elif i == len(cells)-1: m = 0
            l = j - 1
            d = j + 1
            if j == 0 : l = len(cells[0])-1
            elif j == len(cells[0])-1: d = 0
            #Подсчет соседей
            neib = cells[i][l] + cells[i][d] + cells[k][l] + cells[k][j] + cells[k][d] + cells[m][l] + cells[m][j] + cells[m][d]
            if neib == 2 or neib == 3: #Обновление наличия жизни 
                    cells[i][j] = 1 
            else:
                    cells[i][j] = 0 
    #Проверка на условие точки остановы
    for i in data:
        if (sum((sum(i == np.array(cells))))) == 100:
            DO = 0
            break
p.quit()
```
